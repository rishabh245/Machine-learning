```python
import pandas as pd
from sklearn.datasets import load_iris
```

```python
iris1 = load_iris()
iris = pd.DataFrame(iris1.data)
iris.head()
```

```python
print(iris1.feature_names)
```

```python
df = iris.copy()
df.columns = [i for i in iris1.feature_names]
```

```python
print(df.shape)
print(df.dtypes)
```

```python
print(iris1.target_names)
target = []
for i in iris1.target:
    if i==0:
        target.append('setosa')
    elif i==1:
        target.append('versicolor')
    else:
        target.append('virginica')
```

```python
iris['flower_type'] = target
iris.head()
```

```python
iris[iris.flower_type == 'setosa'].describe()
```
