## Customizing graph

```python
import numpy as np
x = np.arange(0,5,0.1)
y = x ** 3
y2 = x ** 2
#plt.plot(x,y,color = 'blue' , marker = 'o')
plt.plot(x,y,color = 'blue' , linewidth = 3 , label = 'x^3')#label is used by legend()
plt.plot(x,y2,color = 'green' , linewidth = 3 ,label = 'x^2')
#plt.axis((0,10,0,150)) #takes a tuple of 4 (Range of x asis and y axis)
plt.grid()
plt.legend()
plt.text(2,80 ,"text" , fontsize = 12) #to add a text at a particular point
plt.xlabel('x')
plt.ylabel('y')
plt.title("Matplotlib demo")
plt.show()
```

## Pie graph

```python
sizes = [4,2,6,3]
colors = ['blue' ,'red' ,'green' ,'violet']
labels = ['A' ,'B' ,'C' ,'D']
explode = [0.1,0,0,0]
plt.pie(sizes,colors=colors,labels = labels,autopct = "%.1f%%" ,explode = explode,counterclock=False,startangle=100)
plt.axis('equal')
plt.show()
```
